```python
import random
import tkinter as tk
from tkinter import ttk

# Define the list of games
games = ['Text Twist', 'Word Boggle', 'Bookworm', 'Flip Words', 'Jig Words']

# Define a list of words for each level
level1 = ['cat', 'dog', 'sun', 'pen', 'cup']
level2 = ['tree', 'book', 'bird', 'fish', 'rose']
level3 = ['cake', 'juice', 'chair', 'lamp', 'clock']
level4 = ['guitar', 'banana', 'orange', 'pizza', 'shirt']
level5 = ['elephant', 'computer', 'pencil', 'glasses', 'shampoo']

# Define a function to display the game instructions
def display_instructions():
    instructions = 'Welcome to Text Twist!\n\nYou will be given a jumbled word, and you have to guess the correct word.\nTo submit your answer, type the word and press enter.\nYou have 3 chances to guess the word correctly.\n\nGood luck!'
    return instructions

# Define a function to jumble a word
def jumble_word(word):
    jumbled_word = ''.join(random.sample(word, len(word)))
    return jumbled_word

# Define a function to play the game
def play_text_twist():
    global word_index, words, word
    # Create a new window for the game
    game_window = tk.Tk()
    game_window.title('Text Twist')
    game_window.geometry('600x600')
    game_window.resizable(False, False)
    game_window.configure(background='#f5f5f5')

    # Add a title label
    title_label = tk.Label(game_window, text='Text Twist', font=('Helvetica', 24), bg='#f5f5f5')
    title_label.pack(pady=10)

    # Add a label for the game instructions
    instructions_label = tk.Label(game_window, text=display_instructions(), font=('Helvetica', 12), bg='#f5f5f5', justify='left')
    instructions_label.pack(padx=20, pady=10)

    # Add a frame for the game
    game_frame = ttk.Frame(game_window)
    game_frame.pack(pady=20)

    # Add a label for the jumbled word
    word_label = tk.Label(game_frame, text='', font=('Helvetica', 18))
    word_label.pack()

    # Add an entry for the user's guess
    guess_entry = ttk.Entry(game_frame, font=('Helvetica', 12), width=20)
    guess_entry.pack(pady=10)

    # Add a label for the feedback message
    feedback_label = tk.Label(game_frame, text='', font=('Helvetica', 12))
    feedback_label.pack()

    # Define a function to check the user's guess
    def check_guess():
        # Get the user's guess
        guess = guess_entry.get().lower()

        # Check if the guess is correct
        if guess == word:
            feedback_label.configure(text='Correct!', fg='green')
            next_button.pack(pady=10)
            guess_button.config(state='disabled')
        else:
            feedback_label.configure(text='Incorrect. Try again.', fg='red')

    # Add a button to submit the guess
    guess_button = ttk.Button(game_frame, text='Submit', command=check_guess)
    guess_button.pack(pady=10)

    # Add a button to move on to the next word
    def next_word():
        global word_index, words, level_index, levels, level_complete, word
        guess_entry.config(state='normal')
        guess_button.config(state='normal')
        next_button.pack_forget()
        if word_index < len(words) - 1:
            word_index += 1
            word = words[word_index]
            jumbled_word = jumble_word(word)
            word_label.configure(text=jumbled_word)
            guess_entry.delete(0, tk.END)
            feedback_label.configure(text='')
            guess_entry.focus()
        else:
            level_index += 1
            if level_index < len(levels):
                level_complete = True
                words = levels[level_index]
                random.shuffle(words)
                word_index = 0
                word = words[word_index]
                jumbled_word = jumble_word(word)
                word_label.configure(text=jumbled_word)
                guess_entry.delete(0, tk.END)
                feedback_label.configure(text='')
                level_label.configure(text=f'Level {level_index+1}')
            else:
                game_window.destroy()

    next_button = ttk.Button(game_frame, text='Next', command=next_word)
    # Add a label for the current level
    level_index = 0
    levels = [level1, level2, level3, level4, level5]
    level_label = tk.Label(game_frame, text=f'Level {level_index+1}', font=('Helvetica', 12))
    level_label.pack()

    # Define a function to start the game
    def start_game():
        global word_index, words, word
        # Get the current word and jumble it
        words = levels[level_index]
        random.shuffle(words)
        word_index = 0
        word = words[word_index]
        jumbled_word = jumble_word(word)
        # Update the word label
        word_label.configure(text=jumbled_word)
        # Clear the guess entry and feedback label
        guess_entry.delete(0, tk.END)
        feedback_label.configure(text='')
        # Reset the state of the guess button and entry
        guess_button.config(state='normal')
        guess_entry.config(state='normal')
        # Hide the next button
        next_button.pack_forget()
        # Focus on the guess entry
        guess_entry.focus()

    # Start the game
    level_complete = False
    start_game()

    # Start the main loop
    game_window.mainloop()
    
# Code the Word Boggle Game

def display_instructions():
    instructions = 'Welcome to Word Boggle!\n\nYou will be given a 4x4 grid of letters, and you have to form as many words as possible using the letters.\nTo submit your answer, type the word and press enter.\nYou have 3 minutes to form as many words as possible.\n\nGood luck!'
    return instructions

def generate_grid():
    grid = [[random.choice('abcdefghijklmnopqrstuvwxyz') for _ in range(4)] for _ in range(4)]
    return grid

def play_word_boggle():
    game_window = tk.Tk()
    game_window.title('Word Boggle')
    game_window.geometry('600x600')
    game_window.resizable(False, False)
    game_window.configure(background='#f5f5f5')

    title_label = tk.Label(game_window, text='Word Boggle', font=('Helvetica', 24), bg='#f5f5f5')
    title_label.pack(pady=10)

    instructions_label = tk.Label(game_window, text=display_instructions(), font=('Helvetica', 12), bg='#f5f5f5', justify='left')
    instructions_label.pack(padx=20, pady=10)

    game_frame = ttk.Frame(game_window)
    game_frame.pack(pady=20)

    grid_label = tk.Label(game_frame, text='', font=('Helvetica', 18))
    grid_label.pack()

    answer_entry = ttk.Entry(game_frame, font=('Helvetica', 12), width=20)
    answer_entry.pack(pady=10)

    score_label = tk.Label(game_frame, text='Score: 0', font=('Helvetica', 12))
    score_label.pack()

    # Define this list before using it
    level1 = []
    valid_words_label = tk.Label(game_frame, text='Valid words: ' + ', '.join(level1), font=('Helvetica', 12))
    valid_words_label.pack()

    def update_grid():
        nonlocal grid_label
        grid = generate_grid()
        grid_str = ''
        for row in grid:
            grid_str += ' '.join(row) + '\n'
        grid_label.configure(text=grid_str)

    def check_answer():
        nonlocal score_label, valid_words_label, level1
        answer = answer_entry.get().lower()
        answer_entry.delete(0, tk.END)

        if answer in level1:
            score_label.configure(text=f'Score: {int(score_label.cget("text").split(":")[1])+1}')
            valid_words = valid_words_label.cget('text').split(': ')[1].split(', ')
            valid_words.append(answer)
            valid_words_label.configure(text=f'Valid words: {", ".join(valid_words)}')

    submit_button = ttk.Button(game_frame, text='Submit', command=check_answer)
    submit_button.pack(pady=10)

    time_label = tk.Label(game_frame, text='Time: 3:00', font=('Helvetica', 12))
    time_label.pack(pady=10)

    def update_time():
        nonlocal time_label
        time = time_label.cget('text').split(':')
        minutes = int(time[1].split(' ')[1])
        seconds = int(time[2])
        if minutes == 0 and seconds == 0:
            game_window.destroy()
        elif seconds == 0:
            minutes -= 1
            seconds = 59
        else:
            seconds -= 1
        time_label.configure(text=f'Time: {minutes:02}:{seconds:02}')
        game_window.after(1000, update_time)

    # Update the grid and start the timer
    update_grid()
    update_time()

    # Start the main loop
    game_window.mainloop()

# Call the function to start the game
play_word_boggle()



# Code the Bookworm Game

# Code the Flip Words Game

# Code the Jig Words Game

# Define a function to ask the user to choose a game
def play_game(game):
    if game == 'Text Twist':
        play_text_twist()
    elif game == 'Word Boggle':
        word_boggle()
    elif game == 'Bookworm':
        bookworm()
    elif game == 'Flip Words':
        flip_words()
    elif game == 'Jig Words':
        jig_words()

# Define a function to display the menu
def display_menu():
    # Create a new window
    menu_window = tk.Tk()
    menu_window.title('Memory Game')
    menu_window.geometry('400x500')
    menu_window.resizable(False, False)
    menu_window.configure(background='#f5f5f5')

    # Add a title label
    title_label = tk.Label(menu_window, text='Memory Game', font=('Helvetica', 24), bg='#f5f5f5')
    title_label.pack(pady=10)

    # Add a label for the creator
    creator_label = tk.Label(menu_window, text='Created by Joshua Bumanlag', font=('Helvetica', 10), bg='#f5f5f5')
    creator_label.pack()

    # Add a frame for the game options
    game_frame = ttk.Frame(menu_window)
    game_frame.pack(pady=20)

    # Add a label for the game options
    game_label = tk.Label(game_frame, text='Choose a game to play:', font=('Helvetica', 14))
    game_label.pack()

    # Add a button for each game
    for i, game in enumerate(games):
        button = ttk.Button(game_frame, text=game, command=lambda game=game: play_game(game))
        button.pack(fill='x', padx=20, pady=5)

    # Add an exit button
    exit_button = ttk.Button(menu_window, text='Exit', command=menu_window.quit)
    exit_button.pack(pady=10)

    # Start the main loop
    menu_window.mainloop()

# Call the display_menu function
display_menu()
```
